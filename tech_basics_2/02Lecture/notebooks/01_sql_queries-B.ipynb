```python
# remove where description contains ? in description

query = """
        select t1.*,
               t2.Description
        from online_retail_history as t1
        join stock_description as t2 on t1.StockCode = t2.StockCode
        /* remove stocks with ? in the description */
        where t2.Description not like '%?%'
"""

pd.read_sql(query, conn)
```

```python
# question: any other data checks we could carry out and clean data based on?
# from table above, can see missing customer ids - any other descriptions we want to remove?


query = """
        select t1.*,
               t2.Description
        from online_retail_history as t1
        join stock_description as t2 on t1.StockCode = t2.StockCode
        /* remove stocks with ? in the description */
        where t2.Description not like '%?%'
            /* remove where customer id are missing */
            and t1.CustomerID is not null
            /* remove where StockCode is POST*/
            and t1.StockCode <> 'POST'
"""

pd.read_sql(query, conn)
```

### Activity: What are the Top Ten Best Selling Products?

```python
# query to fetch the top 10 best selling products

query = """
        select t2.Description,
               sum(t1.Quantity) as number_purchased
        from online_retail_history as t1
        join stock_description as t2 on t1.StockCode = t2.StockCode
        group by t2.Description
        order by number_purchased desc
        limit 10
"""

pd.read_sql(query, conn)
```

```python
# query to fetch the top 10 best selling products with filtering
query = """
        select t2.Description,
               sum(t1.Quantity) as number_purchases
        from online_retail_history as t1
        join stock_description as t2 on t1.StockCode = t2.StockCode
        where t2.Description <> '?'
        group by t2.Description
        order by number_purchases desc
        limit 10
"""

pd.read_sql(query, conn)
```
